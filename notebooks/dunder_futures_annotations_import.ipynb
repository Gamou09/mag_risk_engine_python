# Usage of __future__ (reads dunder futures) import annotations

- TL;DR: it tells python don’t evaluate type hints immediately — keep them as strings and resolve them later.”
- Become default after python 3.11+ - so only really necessary if older version or for consistency 

# Why this is a big deal (real benefits)
- Forward references “just work”
- Faster imports because Type Hints can Import modules, Trigger expensive code, Cause circular imports. Deferring evaluation avoids that !
- Fewer circular import headaches
- Future-proof typing syntax

```python

def g(x: int) -> list[int]:
    pass

g.__annotations__
```

```python
from __future__ import annotations

def f(x: int) -> list[int]:
    pass

f.__annotations__
```
